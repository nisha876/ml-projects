```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_set=pd.read_csv("salary_data.csv")
```

```python
x=data_set.iloc[:,:-1].values
y=data_set.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=1/3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train, y_train)
```

```python
y_pred=regressor.predict(x_test)
x_pred=regressor.predict(x_train)
```

```python
plt.scatter(x_train,y_train,color='green')
plt.plot(x_train, x_pred, color='red')
plt.title("Salary vs Experience (Training Dataset)")
plt.xlabel("Years of experience")
plt.ylabel("Salary(in Ruppes)")
plt.show()
```

```python
plt.scatter(x_test, y_test, color='red')
plt.plot(x_train, x_pred, color='blue')
plt.title("Salary vs Experience(Test Dataset)")
plt.xlabel("years of experience")
plt.ylabel("Salary(in Ruppes)")
plt.show()
```

```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import precision_score, recall_score, average_precision_score

# Load the dataset
dataset = pd.read_csv('Ratings.csv')  # Assuming you have a CSV file with book ratings

# Preprocess the data if needed (e.g., handle missing values, normalize features)

# Split the data into features (X) and target (y)
X = dataset.drop('ISBN', axis=1)  # Assuming 'BookID' is not a feature for recommendation
y = dataset['ISBN']

# Create an instance of the KNN model
knn_model = NearestNeighbors(n_neighbors=5, algorithm='auto')  # Adjust the number of neighbors as needed

# Fit the model to the data
knn_model.fit(X)

# Make recommendations for all books in the dataset
all_recommendations = []
for book_index in range(len(X)):
    distances, indices = knn_model.kneighbors(X.iloc[book_index, :].values.reshape(1, -1))
    recommended_books = y.iloc[indices[0]].values
    all_recommendations.append(recommended_books)

# Flatten the ground truth and recommendations
flattened_ground_truth = y.values.flatten()  # Flatten the ground truth
flattened_recommendations = [rec for sublist in all_recommendations for rec in sublist]  # Flatten the recommendations

# Calculate precision and recall
precision = precision_score(flattened_ground_truth, flattened_recommendations, average='micro')
recall = recall_score(flattened_ground_truth, flattened_recommendations, average='micro')

print("Precision:", precision)
print("Recall:", recall)

# Calculate mean average precision (MAP)
average_precision = average_precision_score(
    [1 if rec in flattened_ground_truth else 0 for rec in flattened_recommendations],
    [1 if rec in flattened_recommendations else 0 for rec in flattened_ground_truth]
)
print("Mean Average Precision (MAP):", average_precision)
```
