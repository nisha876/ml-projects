```python
data = pd.read_csv('C:\\Users\\Shubhangam\\Desktop\\Yoshops\\Project task\\Priject task 2\\Csv files\\orders_2016-2020_Dataset.csv')
data.head()
```

```python
dt_py = data[['Order #','Payment Method']]
dt_py
```

```python
dt_py.dropna(inplace=True)
```

```python
dt_py['Method'] = dt_py['Payment Method'].str.split(pat=" ",expand =True)[0]
```

```python
dt_py['Method'].value_counts()
```

```python
w = dt_py.groupby('Method').count().sort_values('Order #', ascending= False).reset_index()
w
plt.gcf()
sns.barplot(data=w, x= 'Order #', y='Payment Method')
plt.xlabel('Method of Paying')
plt.ylabel('Order Number')
plt.show()
```

Enter-3  to see the analysis of Top Consumer States of India

```python
dt_st = data[['Order #', 'Billing State']]
dt_st
```

```python
#dropping na values
dt_st.dropna(inplace=True)
```

```python
#counting the top states
top_st = dt_st.groupby('Billing State').count().sort_values('Order #', ascending =False).reset_index()
top_st
```

```python
plt.gcf()
sns.barplot(data=top_st.head(15), y= 'Billing State', x='Order #')
plt.xlabel('States')
plt.ylabel('Purchase_count')
plt.title('Purchase count by states')
plt.show()
```

Enter 4 to see the analysis of Top Consumer Cities of India

```python
dt_ct = data[['Order #', 'Billing City']]
dt_ct.head()
```

```python
top_cts = dt_ct.groupby('Billing City').count().sort_values('Order #',ascending=False).reset_index()
top_cts
```

```python
#plotting Graph
plt.gcf()
sns.barplot(data=top_cts.head(15), y='Billing City', x='Order #')
plt.xlabel('Cities')
plt.ylabel('Purchase Count')
plt.title('Purchase Count By Cities')
plt.show()
```

Enter 7 to see the analysis of Number of orders per month per year

```python
#creating data for time and date
time_data = data[['Order #','Order Date and Time Stamp']]
time_data.head()
```

```python
#converting into timestamp for better workflow
time_data['Order Date and Time Stamp'] = pd.to_datetime(time_data['Order Date and Time Stamp'])
time_data['Order Date and Time Stamp']
```

```python
time_data['Order Date and Time Stamp'].dtypes
```

```python
time_data['year'] = pd.to_datetime(time_data['Order Date and Time Stamp']).dt.year
time_data['month'] = pd.to_datetime(time_data['Order Date and Time Stamp']).dt.month
time_data['month_name'] = pd.to_datetime(time_data['Order Date and Time Stamp']).dt.month_name()
time_data['day'] = pd.to_datetime(time_data['Order Date and Time Stamp']).dt.day
time_data['day_name'] = pd.to_datetime(time_data['Order Date and Time Stamp']).dt.day_name()
time_data['hour'] = pd.to_datetime(time_data['Order Date and Time Stamp']).dt.hour
```

```python
time_data['year']
```

```python
time_data['month']
```

```python
time_data['month_name']
```

```python
time_data['day']
```

```python
time_data['day_name']
```

```python
time_data['hour']
```

```python
#grouping time data per year
year_count = time_data.groupby('year').count().sort_values('Order #', ascending=False).reset_index()
year_count
```

```python
plt.gcf()
sns.barplot(data=year_count, x='year', y='Order #')
plt.xlabel('Years')
plt.ylabel('purhase product count')
plt.title('Product purchase count yearwise')
```

```python
#grouping time data per month
month_count = time_data.groupby('month_name').count().sort_values('Order #', ascending=False).reset_index()
month_count
```

```python
plt.gcf()
sns.barplot(data=month_count, x= 'month_name', y='Order #')
plt.xlabel('Month')
plt.ylabel('Product purchase count')
plt.title('Product purchaee monthwise')
plt.xticks(rotation = 90)
plt.show()
```

```python
#grouping time_data per year and month
per_year_per_month = time_data.groupby(['year', 'month_name']).count().sort_values('Order #', ascending=False).reset_index()
per_year_per_month

```

```python
plt.gcf()
sns.barplot(per_year_per_month, x='month_name',y='Order #', hue='year')
plt.xlabel('Month')
plt.ylabel('Purchase per year per month')
plt.xticks(rotation = 90)
plt.show()
```

```python
x = per_year_per_month[['year', 'month_name', 'Order #']].sort_values('Order #', ascending=False).reset_index()
x

plt.gcf()
sns.barplot(data= x, x='year', y='Order #',hue='month_name')
plt.show()
```

Enter 9 to see the analysis of Number of Orders Across Parts of a Day

```python
z= time_data.groupby(['day_name', 'hour','year', 'month_name']).count().sort_values('Order #', ascending=False).reset_index()
plt.gcf()
sns.barplot(data=z, x='day_name', y='hour', hue= 'month_name')
plt.xlabel('Day Name')
plt.ylabel('Hour')
plt.title('Number of orders across parts of a day')
matplotlib.rcParams['figure.figsize'] = (20,15)
matplotlib.rcParams['font.size']= 50
plt.show()
```
